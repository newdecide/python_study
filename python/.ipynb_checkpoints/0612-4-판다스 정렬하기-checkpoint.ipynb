```python
# 1.구구단 인덱스 거꾸로 생성하라
import numpy as np
import pandas as pd
se2 = pd.Series(np.arange(10)*9)
se2.sort_index(ascending=False)
```

```python
# 3.7단, 8단 , 9단을 df2에 추가하라
dan7 = pd.Series(np.arange(10)*7)
dan8 = pd.Series(np.arange(10)*8)
dan9 = pd.Series(np.arange(10)*9)

df2 = pd.DataFrame({
    '7단' : dan7,
    '8단' : dan8,
    '9단' : dan9
})
df2
```

```python
# 칼럼기준 디폴트 오름차순 정렬
df2.sort_index(axis=1)
```

```python
# 칼럼 기준으로 내림차순으로 보여줘라
df2.sort_index(axis=1, ascending=False)
```
