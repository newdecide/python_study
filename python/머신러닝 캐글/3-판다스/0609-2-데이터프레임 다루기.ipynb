```python
import pandas as pd
import numpy as np

# 딕셔너리 타입
data = {
    'id': [1,2,3],
    'name': ['Oh','Lee','Kim'],    
    'age': [10,20,30],
    'assets': [150.4,20.1,325.23],
    'job': ['student','CEO','Dad'],
}
# 데이터 프레임으로 변환
df = pd.DataFrame(data)
df
```

```python
# 데이터프레임 0번 시리즈 출력 타입 시리즈
df.ix[0]
```

```python
# 데이터 프레임 id 칼럼만 출력 시리즈 타입
df['id']
```

```python
# 취미 칼럼 추가 브로드캐스팅으로 책 읽기가 다 들어감
df['취미'] = "책 읽기"  # 추가
```

```python
df
```

```python
# del // drop 삭제

# id가 삭제 됨 axis가 1이면 칼럼기준으로 삭제
df.drop('id', axis =1)
```

```python
df.drop(0)
# 0을 빼면 디폴트로 빠진다.
# 반환은 되지 않는다.
# 0은 세로축 1은 가로축
```

```python
# id칼럼을 인덱스로 옮김
# 인덱스는 임의의 값이 들어있었다.
# id 중복값 제거하기위해 제거 하고 출력한 것
# df에 저장하고 id를 인덱스로 줄 수 있다.
df.index = df['id']
df = df.drop('id', axis=1)
df
```

```python
# 위치를 찾는 용도 0번째 데이터 출력
df.iloc[0] 
```

```python
# 데이터 프레임 시리즈 추가 컬럼이 동일해야 출력된다. 
df.loc[4] =  ['Hong', 60, 2000, '프리랜서', '음악'] 
df
```

```python
df.T # 행과 열을 뒤집어 준다. 영구히 뒤집으려면 df에 담으면 된다. df = df.T
# 뒤집는 경우는 시리즈가 거꾸로 나올때 뒤집는다.
```

```python
# 1.구구단 1단 부터 3단가지 출력하는 데이터 프레임 df1을 생성하라 컬럼 3, 로우 9개
dan1 = pd.Series(list(range(10))) # 시리즈 생성
dan2 = dan1 * 2 # 브로드캐스트 연산
dan3 = dan1 * 3 # 브로드캐스트 연산

data2 = {
    '1단': dan1,
    '2단': dan2,
    '3단' : dan3
} # 시리즈로 묶어준다.
df1 = pd.DataFrame(data2) # 시리즈를 데이터프레임으로 변환
df1
```

```python
# 2.df1의 인덱스를 한글로 바꿔라
df1.index = ['영','일','이','삼','사','오','육','칠','팔','구']
df1
```

```python
# 3.df1에 4단을 추가하시오.
# 인덱스를 한글로 바꾼 경우 넣지 못한다. 숫자일 경우 숫자가 일치하지 않았다. 넘파이로 넣어 줄 수 있다.
dan4 = np.arange(10) * 4
df1['4단'] = dan4
df1
```

```python
# 4.df1 의 1단을 11단으로 수정하라
dan11 = np.arange(10)*11
df1['1단'] = dan11
df1.columns = ['11단', '2단', '3단','4단']
df1
```

```python
# 5.11단을 제거하시오.
df1.drop('11단', axis=1)
```

```python
# 데이터프레임 행렬을 변환
df1.T
```
