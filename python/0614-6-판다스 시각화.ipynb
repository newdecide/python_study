```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# cumsum: 누적합계 np.random.randn: 난수
se = pd.Series(np.random.randn(100).cumsum())
se.plot()
```

```python
# 100행 5열 랜덤값만듦
df = pd.DataFrame(np.random.randn(100,5).cumsum(0),
                 columns = ['a1','a2','a3','a4','a5']
                 )
df.plot()
```

```python
# 세로막대 차트
se.plot(kind='bar')
plt.show()
# 가로막대 차트
se.plot(kind='barh')
plt.show()
```

```python
# 라인차트를 세로막대 차트로 변경
df.plot(kind  = 'bar')
```

```python
# 그룹화된 세로막대 차트
df.plot(kind = 'bar', stacked=True)
```

```python
se.hist(bins=10) # 히스토그램
se.plot(kind='kde') # 밀도 측정
plt.show()
```

```python
se.hist(bins=10, normed=True) # 히스토그램
se.plot(kind='kde') # 밀도 측정
plt.show()
```

```python
from sklearn import datasets

iris = datasets.load_iris()
irisdt = pd.DataFrame(iris['data'], columns = iris['feature_names'])
irisdt

# 꽃에는 4가지 특징이 있다.
```

```python
plt.scatter(irisdt['sepal length (cm)'],irisdt['sepal length (cm)'],irisdt['sepal length (cm)'],irisdt['sepal length (cm)'])
```

```python
plt.scatter(irisdt['sepal length (cm)'],irisdt['petal length (cm)'],c=iris['target'])
```

```python
pd.plotting.scatter_matrix(irisdt, diagonal ='kde', alpha=0.3)
plt.show()
```

```python
from sklearn import neighbors
# 부동산 가격 예측
knn = neighbors.KNeighborsClassifier()
knn.fit(iris['data'], iris['target']) # 훈련시키는 작업
knn.score(iris['data'], iris['target']) # 점수
```
