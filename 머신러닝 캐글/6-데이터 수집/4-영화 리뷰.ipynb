```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import time 

review_url = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=161967&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={}"


html = urlopen(review_url.format(200))
bsObj = BeautifulSoup(html.read(), "html.parser")
```

```python
list1 = bsObj.find_all("div", {"class": "score_result"})
list1
```

```python
# 레이블 처리
list2 = list1[0].findAll('li')
len(list2)
```

```python
import pandas as pd
count = 0
review_dict = dict()

for li in list2:
    review_dict[count] = {"평점": int(li.em.text), '리뷰': li.p.text}
    count += 1
    

pd.DataFrame(review_dict).T
```
