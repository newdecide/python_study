```python
import pandas as pd
df = pd.read_csv('ratings_train.txt', sep='\t')
df
```

```python
df['label'].value_counts()
```

```python
import re # 정규표현식
df = df.fillna(' ')
df['document'].apply(lambda x:re.sub(r'\d+', '', x)) # 기능을 다 실행해준다.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from konlpy.tag import Twitter

tw = Twitter() # 토크나이저(토큰화하는 객체) 생성
tfidf_vect = TfidfVectorizer(tokenizer=tw.morphs, ngram_range=(1,2), # 토큰화로 사용할 함수, n_gram 범위(토큰화에 사용될 단어길이)
               min_df=3, max_df=0.9)# 빈도수 최소 3개 이상, 상위 10%는 삭제

tfidf_vect.fit(df['document']) #벡터라이저 훈련
```

```python
# 훈련 데이터에 벡터라이징 적용
tfdif_matrix_train = tfidf_vect.transform(df['document'])
```

```python
clf = LogisticRegression(random_state=0) # 분류자 생성
params = {'C':[1, 3.5, 4.5, 5.5, 10]} # 분류자에 사용할 파라미터 값
grid_cv = GridSearchCV(clf, param_grid = params, cv=3, scoring='accuracy', verbose=1) # 다양한 케이스로 분류자 훈련, 평가:정확도
# accuracy 정확도 정답 맞춘 개수, verbose=1 설명을 이렇게 해달라
grid_cv.fit(tfdif_matrix_train, df['label']) # 분류자 훈련
```

```python
grid_cv.best_score_, grid_cv.best_params_ # 최고성능, 파라미터 확인
```

```python
#중간에 에러나면 창을 밑에서 해주세요. document가 오래걸립니다.
```

```python
case1 = '이 영화 너무 재미있다'
case2 = '이 영화 너무 재미없다.'
case3 = '이것도 영화라고 찍냐'
case1 = tfidf_vect.transform([case1, case2, case3])
grid_cv.best_estimator_.predict(case1)
```
