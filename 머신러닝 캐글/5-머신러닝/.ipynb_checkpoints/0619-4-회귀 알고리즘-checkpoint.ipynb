```python
import pandas as pd
from sklearn.model_selection import train_test_split

path = 'dbs/50_Startups.csv'
df1 = pd.read_csv(path)
df1
```

```python
# x,y구분
# 영업이익 y skajwls 나머지가 x가 된다. 3번은 버림

x = df1.values[:,:3] # 0~2까지
y = df1.values[:,4] # 4번만
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)

len(x_train), len(x_test)
```

```python
result = dict()

from sklearn  import linear_model

reg = linear_model.Ridge() # 선형 많이 쓰는 함수
reg.fit(x_train, y_train)
reg.score(x_train, y_train)
result['Ridge'] = reg.score(x_test, y_test)

reg = linear_model.Lasso() # 
reg.fit(x_train, y_train)
reg.score(x_train, y_train)
result['Lasso'] = reg.score(x_test, y_test)

from sklearn.svm import SVR
reg = SVR()
reg.fit(x_train, y_train)
result['SVR'] = reg.score(x_test, y_test)

from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(x_train, y_train)
result['Tree'] = reg.score(x_test, y_test)

import pandas as pd
pd.Series(result)
```
