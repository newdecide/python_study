```python
from sklearn.datasets import load_boston

boston = load_boston()
```

```python
from sklearn.tree import DecisionTreeClassifier

x = boston['data']
y = boston['target']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
x_test.shape, y_test.shape
```

```python
print(boston.DESCR)
```

```python
result = dict()
model = dict()
```

```python
from sklearn  import linear_model

reg = linear_model.Ridge() # 선형 많이 쓰는 함수
reg.fit(x_train, y_train)
reg.score(x_train, y_train)
result['Ridge'] = reg.score(x_test, y_test)
model["Ridge"] = reg

reg = linear_model.Lasso() # 
reg.fit(x_train, y_train)
reg.score(x_train, y_train)
result['Lasso'] = reg.score(x_test, y_test)
model["Lasso"] = reg

from sklearn.svm import SVR
reg = SVR()
reg.fit(x_train, y_train)
result['SVR'] = reg.score(x_test, y_test)
model["SVR"] = reg

from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(x_train, y_train)
result['Tree'] = reg.score(x_test, y_test)
model["Tree"] = reg

import pandas as pd
pd.Series(result)
```

```python
pred_y = model['Ridge'].predict(x_test)
pred_y
```
